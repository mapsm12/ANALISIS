```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.dates as mdates

# Leer los datos desde un archivo CSV
df = pd.read_excel("./drive/MyDrive/jquispeSerie/DataTemSal  IMARPE21Jun2023.xlsx")
data=df

# Convertir la columna 'Fecha' a formato de fecha
data['Fecha'] = pd.to_datetime(data['Fecha'])

# Obtener los valores de fecha y temperatura como arrays NumPy
fechas = mdates.date2num(data['Fecha'].values)
temperatura = data['TSM'].values

# Definir la función de ajuste (en este caso, una función sinusoidal)
def sinusoidal(x, a, b, c, d):
    return a * np.sin(b * x + c) + d

# Realizar el ajuste de la función a los datos
popt, pcov = curve_fit(sinusoidal, fechas, temperatura)

# Generar valores predichos utilizando los parámetros ajustados
temperatura_predicha = sinusoidal(fechas, *popt)

# Convertir las fechas de números de serie a fechas
fechas_convertidas = mdates.num2date(fechas)

# Configurar la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la serie de tiempo de temperatura
ax.plot(fechas_convertidas, temperatura, label='Temperatura')
ax.plot(fechas_convertidas, temperatura_predicha, label='Ajuste')

# Configurar etiquetas y título
ax.set_xlabel('Fecha')
ax.set_ylabel('Temperatura')
ax.set_title('Serie de Tiempo de Temperatura con Ajuste Sinusoidal')

# Rotar las etiquetas del eje x para mayor legibilidad
plt.xticks(rotation=45)

# Mostrar la leyenda
ax.legend()

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Leer los datos desde un archivo CSV
df = pd.read_csv("./drive/MyDrive/jquispeSerie/TimeSeriesData2023.csv")
df = df[['Fecha', 'Temperatura']]
df = df[df['Temperatura'] > 10]

# Convertir la columna "Fecha" a tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Realizar el promedio por valor de fecha cada 3 días
df_avg = df.resample('D', on='Fecha').mean().ffill().bfill()

# Calcular la anomalía de temperatura (diferencia con el promedio)
df_anomalia = df.set_index('Fecha') - df_avg.reindex(df.set_index('Fecha').index)

# Realizar la interpolación utilizando el método PCHIP
df_interpolated = df_anomalia.interpolate(method='pchip')

# Realizar el análisis de Fourier
y = df_interpolated['Temperatura'].values
n = len(y)  # Número de puntos
dt = 1  # Espaciado temporal (1 día en este caso)

# Aplicar la transformada de Fourier
yf = fft(y)
xf = fftfreq(n, dt)[:n // 2]  # Frecuencias positivas

# Calcular la potencia espectral
power = np.abs(yf[:n // 2]) ** 2

# Crear un gráfico de la potencia espectral
plt.figure(figsize=(12, 6))
plt.plot(xf, power)
plt.xlabel('Frecuencia (1/días)')
plt.ylabel('Potencia Espectral')
plt.title('Análisis de Fourier - Serie de Anomalía de Temperatura')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

# Leer los datos desde un archivo CSV
df = pd.read_excel("./drive/MyDrive/jquispeSerie/DataTemSal  IMARPE21Jun2023.xlsx")

# Obtener los valores de fecha y SSM como arrays NumPy
fechas = pd.to_datetime(df['Fecha'])  # Convertir a formato de fecha
df["Fecha"]= pd.to_datetime(df['Fecha'])  # Convertir a formato de fecha
# SERIE ORIGINAL
plt.plot(df["Fecha"],df["SSM"])
temperatura = df['SSM'].values

# Generar las fechas diarias entre la fecha mínima y máxima
fechas_completas = pd.date_range(start=fechas.min(), end=fechas.max(), freq='D')

# Crear una nueva tabla con las fechas completas
tabla_interpolada = pd.DataFrame({'Fecha': fechas_completas})

# Combinar la tabla interpolada con la tabla original conservando los valores existentes
tabla_interpolada = pd.merge(tabla_interpolada, df, on='Fecha', how='left')

# Interpolación diaria entre los valores existentes
tabla_interpolada['SSM'] = tabla_interpolada['SSM'].interpolate(method='pchip')

# Obtener los valores de fecha y SSM interpolada como arrays NumPy
fechas2 = pd.to_datetime(tabla_interpolada['Fecha'])  # Convertir a formato de fecha
temperatura2 = tabla_interpolada['SSM'].values

# Crear una copia de la serie original
temperatura_reconstruida = np.copy(temperatura2)

# Encontrar los índices de los valores faltantes (NaN)
indices_faltantes = np.isnan(temperatura2)

# Aplicar la transformada de Fourier a la serie de tiempo
transformada = fft(temperatura2)

# Obtener los coeficientes de amplitud y fase de cada armónico
amplitudes = np.abs(transformada)
fases = np.angle(transformada)

# Ordenar los armónicos por amplitud (de mayor a menor)
indices_ordenados = np.argsort(amplitudes)[::-1]
amplitudes_ordenadas = amplitudes[indices_ordenados]
fases_ordenadas = fases[indices_ordenados]

# Reconstruir la serie de tiempo utilizando los 30 primeros armónicos
armonicos_reconstruidos = np.zeros_like(transformada, dtype=np.complex128)
armonicos_reconstruidos[indices_ordenados[:30]] = transformada[indices_ordenados[:30]]

temperatura_reconstruida = ifft(armonicos_reconstruidos).real

# Reemplazar los valores faltantes en la serie original con los valores reconstruidos
temperatura2[indices_faltantes] = temperatura_reconstruida[indices_faltantes]

# Agregar la columna de SSM interpolada a la tabla original
tabla_interpolada['SSM_interpolado'] = temperatura2

# Mostrar los valores NaN en la serie original antes del relleno
valores_nulos = tabla_interpolada.loc[indices_faltantes, ['Fecha', 'SSM']]
print("Valores NaN antes del relleno:")
print(valores_nulos)

# Configurar la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la serie de tiempo original y la reconstruida
ax.plot(fechas2, temperatura2, label='SSM original con valores faltantes reemplazados')
ax.plot(fechas2, temperatura_reconstruida, label='SSM reconstruida')

# Configurar etiquetas y título
ax.set_xlabel('Fecha')
ax.set_ylabel('SSM')
ax.set_title('Serie de Tiempo de SSM con Valores Faltantes Reemplazados')

# Mostrar la leyenda
ax.legend()

# Mostrar la gráfica
plt.show()

# Filtrar los datos entre 2020 y 2021
filtro = (tabla_interpolada["Fecha"] >= "2013-01-01") & (tabla_interpolada["Fecha"] <= "2024-12-31")
tabla_filtrada = tabla_interpolada.loc[filtro]

# Convertir las fechas a objetos de fecha
fechas_filtradas = pd.to_datetime(tabla_filtrada["Fecha"])


plt.figure(figsize=(16, 8))
plt.scatter(fechas_filtradas, tabla_filtrada["SSM_interpolado"], label="Valores Interpolados", alpha=0.7)
plt.scatter(df["Fecha"], df["SSM"], c="red", label="Datos Originales", alpha=0.7)
plt.xlabel('Fecha')
plt.ylabel('SSM')
plt.title('Valores Interpolados y Datos Originales (2020-2021)')
plt.legend()
plt.show()

# Graficar los datos originales y los valores interpolados en el rango de fechas filtrado
plt.figure(figsize=(16, 8))
plt.scatter(fechas_filtradas, tabla_filtrada["SSM_interpolado"], label="Valores Interpolados", alpha=0.7)
plt.scatter(df["Fecha"], df["SSM"], c="red", label="Datos Originales", alpha=0.7)
plt.xlim(pd.to_datetime("2020-01-01"), pd.to_datetime("2020-07-31"))
plt.xlabel('Fecha')
plt.ylabel('SSM')
plt.title('Valores Interpolados y Datos Originales (2020-2021)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense, RepeatVector, TimeDistributed, Activation

# Leer los datos de temperatura de la serie de tiempo
name = "./TimeSeriesDataLIMAAGUASPROFUNDAS.csv"
df_lima = pd.read_csv(name)

# Convertir la columna "Date" a formato datetime
df_lima["Date"] = pd.to_datetime(df_lima["Date"])

# Filtrar los datos para incluir solo los últimos 8 años
fecha_maxima = df_lima["Date"].max()
fecha_minima_entrenamiento = fecha_maxima - pd.DateOffset(years=20)
df_lima = df_lima[(df_lima["Date"] >= fecha_minima_entrenamiento) & (df_lima["Date"] <= fecha_maxima)]

dates = df_lima['Date'].to_numpy()
temperatures = df_lima['Temperature'].to_numpy()

# Preparar los datos para la red neuronal
train_size = int(len(dates) * 0.9)
train_data = temperatures[:train_size]
test_data = temperatures[train_size:]

# Función para crear secuencias de tiempo con desfase
def create_sequences(data, time_steps):
    sequences = []
    for i in range(len(data) - time_steps):
        sequences.append(data[i:i + time_steps])
    return np.array(sequences)

# Hiperparámetros
time_steps = 12  # Número de pasos en el tiempo para cada muestra
n_features = 1  # Univariable, solo una característica (temperatura)
epochs = 100
batch_size = 32

# Crear secuencias de tiempo para entrenamiento y prueba
train_data = create_sequences(train_data, time_steps)
test_data = create_sequences(test_data, time_steps)

# Dividir los datos de entrenamiento y prueba
X_train, y_train = train_data[:, :-1], train_data[:, -1]
X_test, y_test = test_data[:, :-1], test_data[:, -1]

# Redimensionar los datos para LSTM (número de muestras, número de pasos en el tiempo, número de características)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))

# Crear el modelo LSTM con atención
model = Sequential()
model.add(LSTM(32, input_shape=(time_steps-1, n_features), return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.add(Activation('linear'))
model.add(LSTM(32))
model.add(Dense(1))

# Compilación del modelo
model.compile(optimizer='adam', loss='mae')

# Entrenamiento del modelo
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

# Realización de predicción para la extrapolación de 1 año (12 meses)
extrapolation_data = X_test[-1].reshape(1, time_steps-1, n_features)
extrapolation_predictions = []

for _ in range(12):  # Predicción para 1 año (12 meses)
    prediction = model.predict(extrapolation_data)
    extrapolation_predictions.append(prediction[0, 0])
    extrapolation_data = np.append(extrapolation_data[:, 1:, :], prediction[:, np.newaxis, :], axis=1)

# Generar fechas para la extrapolación de 1 año
next_12_months_dates = pd.date_range(start=df_lima["Date"].max() + pd.DateOffset(days=1), periods=12, freq='M')

# Gráfico de las temperaturas reales y la extrapolación de 1 año
plt.plot(df_lima["Date"], temperatures, label='Real')
plt.plot(next_12_months_dates, extrapolation_predictions, label='Extrapolación', color='red', linestyle='--')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.legend()
plt.xticks(rotation=45)
plt.xlim(df_lima["Date"].min(), next_12_months_dates[-1])  # Zoom de meses a 1 año
plt.show()

# Imprimir las fechas de la extrapolación
print("Fecha máxima registrada:", fecha_maxima)
print("Fecha de extrapolación de 1 año:", next_12_months_dates[-1])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('fast')

from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.preprocessing import MinMaxScaler

# Leer los datos de temperatura de la serie de tiempo
name = "./TimeSeriesDataLIMAAGUASPROFUNDAS.csv"
df_lima = pd.read_csv(name)

# Convertir la columna "Date" a formato datetime
df_lima["Date"] = pd.to_datetime(df_lima["Date"])

# Filtrar los datos para incluir solo los últimos 8 años
fecha_maxima = df_lima["Date"].max()
fecha_minima_entrenamiento = fecha_maxima - pd.DateOffset(years=8)
df_lima = df_lima[(df_lima["Date"] >= fecha_minima_entrenamiento) & (df_lima["Date"] <= fecha_maxima)]

# Convertir los datos a una serie de tiempo con fechas como índices
time_series = pd.Series(df_lima['Temperature'].to_numpy(), index=df_lima['Date'])

PASOS = 7

# convert series to supervised learning
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg

# load dataset
values = time_series.values
# ensure all data is float
values = values.astype('float32')
# normalize features
scaler = MinMaxScaler(feature_range=(-1, 1))
values = values.reshape(-1, 1)  # esto lo hacemos porque tenemos 1 sola dimension
scaled = scaler.fit_transform(values)
# frame as supervised learning
reframed = series_to_supervised(scaled, PASOS, 1)

# split into train and test sets
values = reframed.values
n_train_days = len(df_lima) - (30 + PASOS)
train = values[:n_train_days, :]
test = values[n_train_days:, :]
# split into input and outputs
x_train, y_train = train[:, :-1], train[:, -1]
x_val, y_val = test[:, :-1], test[:, -1]
# reshape input to be 3D [samples, timesteps, features]
x_train = x_train.reshape((x_train.shape[0], 1, x_train.shape[1]))
x_val = x_val.reshape((x_val.shape[0], 1, x_val.shape[1]))

def crear_modeloFF():
    model = Sequential() 
    model.add(Dense(PASOS, input_shape=(1, PASOS), activation='tanh'))
    model.add(Flatten())
    model.add(Dense(1, activation='tanh'))
    model.compile(loss='mean_absolute_error', optimizer='Adam', metrics=["mse"])
    model.summary()
    return model

EPOCHS = 40

model = crear_modeloFF()

history = model.fit(x_train, y_train, epochs=EPOCHS, validation_data=(x_val, y_val), batch_size=PASOS)

results = model.predict(x_val)
plt.scatter(range(len(y_val)), y_val, c='g')
plt.scatter(range(len(results)), results, c='r')
plt.title('validate')
plt.show()

# Últimos días disponibles en el conjunto de prueba
ultimosDias = df_lima['Date'].tail(PASOS).values
ultimosDias = pd.to_datetime(ultimosDias).to_pydatetime()
ultimosDias = np.array([d.timestamp() for d in ultimosDias])
ultimosDias = ultimosDias.reshape(-1, 1)
values = df_lima['Temperature'].tail(PASOS).values
values = values.astype('float32')
# normalize features
values = values.reshape(-1, 1)  # esto lo hacemos porque tenemos 1 sola dimension
scaled = scaler.transform(values)
reframed = series_to_supervised(scaled, PASOS, 1)
reframed.drop(reframed.columns[PASOS], axis=1, inplace=True)
values = reframed.values
x_test = values[0, :].reshape((1, 1, PASOS))

results = []
for i in range(365):
    parcial = model.predict(x_test)
    results.append(parcial[0])
    x_test = np.array(results[-PASOS:]).reshape((1, 1, PASOS))

adimen = np.array(results).reshape(-1, 1)
inverted = scaler.inverse_transform(adimen)

# Crear DataFrame con el pronóstico
pronostico_lima = pd.DataFrame(inverted, index=pd.to_datetime(ultimosDias.flatten(), unit='s'))
pronostico_lima.columns = ['pronostico']
pronostico_lima.plot()
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.title('Pronóstico de temperatura en Lima para 1 año')
plt.show()

# Guardar el pronóstico en un archivo CSV
pronostico_lima.to_csv('pronostico_lima.csv')
```
